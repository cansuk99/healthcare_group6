```python
import pandas as pd

df = pd.read_csv("data/processed/diabetes_cleaned_data.csv")

enc_counts = df.groupby("patient_nbr")["encounter_id"].count().reset_index(name="num_encounters")

print("Total patients:", enc_counts.shape[0])
print(enc_counts["num_encounters"].describe())

multi = enc_counts[enc_counts["num_encounters"] > 1]
print("Patients with >1 encounter:", multi.shape[0])



```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(enc_counts["num_encounters"], bins=50, color="steelblue", edgecolor="white")
plt.title("Distribution of Number of Encounters per Patient")
plt.xlabel("Number of Encounters per Patient")
plt.ylabel("Number of Patients")
plt.grid(axis="y", alpha=0.4)

# --- Save plot ---
plt.tight_layout()
plt.savefig("figures/encounter_distribution_per_patient.png", dpi=300)
plt.show()
# proportion of patients who were ever readmitted
# 12 %
df.groupby("patient_nbr")["readmitted_binary"].max().mean()

```
